```python
V = 1000/1000 # m3
T = 293 # K

price = 0.37 # $ per m3
P = 100*1e5 # Pa
M = 16/1000 # kg/mol
R = 8.31 # J/mol/K
```

```python
# PV = m/M*RT
m = P*V*M/R/T
m # kg
```

```python
Vstd = m*R*T/M/1e5
Vstd
```

```python
cost = Vstd*price
cost
```

```python
has_dealer = 10 # dollars available
v_to_sell= has_dealer/price
v_to_sell # m3
```

```python
ro_std = m/Vstd
ro_std
```

```python
m_to_sell = ro_std*v_to_sell
m_to_sell
```

```python
m_remaining = m - m_to_sell
m_remaining
```

```python
# PV = m/M*RT
p = m_remaining/M*R*T
p/1e5
```

```python
# Bg = Vres/Vstd

import matplotlib.pyplot as plt
import numpy as np

def Bg(p):
    pstd = 1e5
    return pstd/p
```

```python
ps = np.linspace(1, 100, 100)*1e5
bgs = Bg(ps)
plt.plot(ps, bgs)
plt.xlabel('p, Pa')
plt.ylabel('bg')
```

```python
import numpy as np

def P(vstd_sell):
    # v = Bg(p0)*vstd_sell
    m_sell = ro_std*vstd_sell
    m_remaining = m - m_sell
    p1 = R*T/M/V*m_remaining
    return p1

# vs = np.linspace(1, 100, 100)
ps = []
vs = list(range(1, 100))
for v in vs:
    ps.append(P(v)/1e5)


fig, ax = plt.subplots()
ax.grid()
ax.set_xlabel('v, m3')
ax.set_ylabel('p, Pa')
plt.plot(vs, ps)
```
